```python
# https://towardsdatascience.com/building-our-first-neural-network-in-keras-bdc8abbc17f5

#Dependencies
import numpy as np
import pandas as pd
#dataset import
dataset = pd.read_csv('datasets/train.csv') #You need to change #directory accordingly
dataset.head(10) #Return 10 rows of data
```

```python
#Changing pandas dataframe to numpy array
X = dataset.iloc[:,:20].values
y = dataset.iloc[:,20:21].values
```

```python
X
```

```python
# Normalization is a technique used to change the values of an array to a common scale,
# without distorting differences in the ranges of values.

#Normalizing the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# One hot encoding is a process to convert integer [(I think it works with strings too.)] classes into binary values.

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
y = ohe.fit_transform(y).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
#Dependencies
import keras
from keras.models import Sequential
from keras.layers import Dense

# Neural network
model = Sequential()
model.add(Dense(16, input_dim=20, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(4, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=64)
```
